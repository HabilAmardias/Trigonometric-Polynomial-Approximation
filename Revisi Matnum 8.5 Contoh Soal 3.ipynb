```python
import math
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**4-3*x**3+2*x**2-math.tan(x*(x-2))
```

```python
m=5
pi=math.pi
x=[]
for j in range(0,2*m):
    element=(j/5)
    x.append(element)
print(x)
```

```python
z=[]
for i in x:
    element=pi*(i-1)
    z.append(element)
print(z)
```

```python
y=[f(1+(j/pi)) for j in z]
print(y)
```

```python
n=3
a=[]
b=[]
```

```python
def a0(Y):
    number=0
    for i in range(0,2*m):
        number=number+((1/m)*Y[i])
    return number
a.append(a0(y))
print(a)
```

```python
def sumasi_ak(k,X,Y):
    state=0
    for i in range(0,2*m):
        state=state+((1/m)*Y[i]*math.cos(k*X[i]))
    return state
```

```python
for k in range(1,n+1):
    a.append(sumasi_ak(k,z,y))
print(a)
```

```python
def sumasi_bk(k,X,Y):
    state=0
    for i in range(0,2*m):
        state=state+((1/m)*Y[i]*math.sin(k*X[i]))
    return state
```

```python
for k in range(0,n):
    b.append(sumasi_bk(k,z,y))
print(b)
```

```python
def g(X,A,B):
    Sn=(A[0]/2)+A[n]*math.cos(n*X)
    for k in range(1,n):
        Sn=Sn+A[k]*math.cos(k*X)+B[k]*math.sin(k*X)
    return Sn
```

```python
S=[g(i,a,b) for i in z]
print(S)
```

```python
def least_square_error(X,Y):
    E=0
    for j in range(0,2*m):
        E=E+(X[j]-Y[j])**2
    return E
```

```python
print(least_square_error(y,S))
```

```python
fig,ax=plt.subplots(1,1)
ax.plot(z,y,label='f(x)')
ax.plot(z,S,label='S_3(x)')
ax.legend()
ax.set_title('Graph fungsi f(x) dan polinomial S_3(x)')
plt.show()
```
